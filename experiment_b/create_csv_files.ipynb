## Setup dataframe FIJI

With Fiji's plugin TrackMate the spots' data can be exported as a csv file.

```python
df_original = pd.read_csv(os.path.join(wdir, 'data_fiji/spots.csv'), skiprows=[1, 2, 3])
```

```python
df = df_original.copy()
```

The spots are sorted first by the track they are in and then by the frame, such that the final dataframe is ordered by blocks, one for each spot, with its data sequentially ordered.

```python
df = df.sort_values(by=['TRACK_ID', 'FRAME'], ascending=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

## Text file with all data

A csv file is generated with the dataframe's data.

```python
delta_t = 16/60 # 16 minutes between frames, to hours
resolution = 1.5 # 1,5 microns per pixel
folder_name = 'data_extracted'
file_name = 'velocities_data.csv'
```

```python
with open(os.path.join(wdir, folder_name, file_name), 'w') as writefile:
  writefile.write("frame,pos_x,pos_y,vel_x,vel_y,diff_t,track_id\n")

  # Iterate over the tracks and generate their dataframe.
  for i in range(0, len(df.TRACK_ID.unique())):
    df_filtered = df[df['TRACK_ID'] == i]
    df_filtered = df_filtered.reset_index(drop=True)

    # Iterate over the frames (except the last, velocities are undefined here) in the track.
    for j in range(0, len(df_filtered)-1):
      frame = df_filtered['FRAME'][j]
      pos_x = df_filtered['POSITION_X'][j]
      pos_y = df_filtered['POSITION_Y'][j]
      vel_x = (resolution/delta_t) * (df_filtered['POSITION_X'][j + 1] - df_filtered['POSITION_X'][j])
      vel_y = - (resolution/delta_t) * (df_filtered['POSITION_Y'][j + 1] - df_filtered['POSITION_Y'][j]) # Fiji inverts y-axis
      diff_t = df_filtered['FRAME'][j + 1] - df_filtered['FRAME'][j]
      track_id = df_filtered['TRACK_ID'][j]

      if diff_t == 2:
        continue

      writefile.write(f"{frame},{pos_x},{pos_y},{vel_x},{vel_y},{diff_t},{track_id}\n")
```

## One text file per frame

From the previous file, one csv file is generated for each frame.

```python
n_frames = 29
```

```python
df_frames = pd.read_csv(os.path.join(wdir, folder_name, file_name))
```

```python
# Total frames: 30, indexed from 0 to 29.
# Iterate from frame 0 to frame 28, since the velocity is undefined for frame 29.

for frame_idx in range(0, n_frames):
  frame_file_name = 'frame' + str(int(frame_idx)).zfill(3) + '.csv'
  df_filtered = df_frames[df_frames['frame'] == frame_idx]
  df_filtered = df_filtered.reset_index(drop=True)

  with open(os.path.join(wdir, folder_name, frame_file_name), 'w') as writefile:
    writefile.write("pos_x,pos_y,vel_x,vel_y,diff_t\n")

    for j in range(0, len(df_filtered)):
      pos_x = df_filtered['pos_x'][j]
      pos_y = df_filtered['pos_y'][j]
      vel_x = df_filtered['vel_x'][j]
      vel_y = df_filtered['vel_y'][j]
      diff_t = df_filtered['diff_t'][j]

      writefile.write(f"{pos_x},{pos_y},{vel_x},{vel_y},{diff_t}\n")
```
